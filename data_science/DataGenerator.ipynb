```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import random
```

```python
special_challenge = ['Yes', 'No']
student_type =["UK/Home student", "International student", "EU student", "Exchange or visiting student", "Distance/online learner", "prefer not to say"]
#the input will be replaced by the different nationalities in the EU and international students to give a real life dataset
nationality_EU = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czech Republic (Czechia)" ,"Denmark", "Estonia", "Finland", "France", "prefer not to say"]
nationality_international = ["India", "Bangladesh", "Sri Lanka", "Nepal", "Australia", "New Zealand", "United States", "Canada", "Malaysia", "Indonesia", "prefer not to say"]
uk_country =["England", "Scotland", "Wales", "Northern Ireland"]

levelofstudy = ["Undergraduate", "Postgraduate Taught", "Postgraduate Research", "prefer not to say"]
year_duration = [1, 2, 3, 4, 5, "more"]
gender_identity = ["Man", "woman", "Non-binary", "prefer to self-describe", "prefer not to say"]
accomodation_type = ["On-campus (university-managed)", "Off-campus (university-managed)", "private student halls", "private rented", "Family/home"]
funding = ["funded by family", "education loans", "part-time job/internships", "staying with family", "not mentioned above (input)", "prefer not to say"]

parttime_income =[] #append prefer not to say
other_income =[] #staying with family,prefer not to say.

age =["Under 18","18 -23","24 -29","30 -34","35-40", "41 and above", "prefer not to say"]
cost_living = ["staying with family", "less than £600", "£600-£800", "£801-£1000", "£1001-£1200","£1201-£1500","over £1500","prefer not to say"]
groceries =["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
housing_utility =["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
internet_and_communication =["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
# mode_of_travel =["walking", "bicycle", "motor cycle", "car", "public transport", "prefer not to say"]
# travel_university =["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
travel_Bills=["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
entertainment=["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
clothes_and_shopping = ["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
medicine_spending = ["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
other_things =["staying with family", "less than £100", "£101- £200", "£201- £300", "£301- £400", "£401- £500", "over £500", "not spending", "prefer not to say"]
catering = ["staying with family", "self-catered", "catered", "prefer not to say"]
type_of_room =["staying with family", "standard", "En-suite", "Studio", "Couples"]
rent = ["staying with family", "less than £600", "£600-£800", "£801-£1000", "£1001-£1200", "£1201-£1500", "over £1500", "prefer not to say"]
manage = ["Very well", "Fairly well", "Sometimes difficult", "often difficult", "very difficult", "prefer not to say"]
sharing = ["Very well", "Fairly well", "Sometimes difficult", "often difficult", "very difficult", "prefer not to say"]
```

```python
n = 100
np.random.seed(56)
sttype = np.random.choice(student_type, n)
funding = np.random.choice(funding, n)
nationality =[]
partime=[]
famfund =[]
#genrating students nationality randomly
for i in sttype:
    if i == "International student":
        nationality.append(random.choice(nationality_international))
    elif i == "EU student":
        nationality.append(random.choice(nationality_EU))
    elif i == "prefer not to say":
        nationality.append("prefer not to say")
    elif i == "UK/Home student":
        nationality.append(random.choice(uk_country))
    else:
        nationality.append(random.choice(random.choice([uk_country,nationality_international, nationality_EU])))

#fund / income is genrated for part-time job/internships
for i in funding:
    if i == "part-time job/internships":
        dum =np.random.uniform(low= 340, high=915)
        partime.append(dum)
    else:
        partime.append(0)

# fund / income is genrated for funded by family/education loans
for i in funding:
    if i == "funded by family" or i == "education loans" :
        jum = np.random.uniform(low= 340, high=915)
        famfund.append(jum)
    else:
        famfund.append("0")

# students data is created as sample data
studentdata ={
    "special_challenge": np.random.choice(special_challenge, n),
    "student_type": sttype,
    #the nationality 
    "nationality": np.array(nationality),
    "levelofstudy": np.random.choice(levelofstudy,n),
    "year_duration": np.random.choice(year_duration,n),
    "gender_identity": np.random.choice(gender_identity,n),
    "accomodation_type": np.random.choice(accomodation_type,n),
    "funding": funding,
    "parttime_income": np.array(partime),
    "other_income": np.array(famfund),
    "age":np.random.choice(age,n),
    "cost_living": np.random.choice(cost_living,n),
    "groceries": np.random.choice(groceries,n),
    "housing_utility": np.random.choice(housing_utility,n),
    "internet_and_communication": np.random.choice(internet_and_communication,n),
    # "mode_of_travel": np.random.choice(mode_of_travel,n),
    # "travel_university": np.random.choice(travel_university,n),
    "travel_Bills": np.random.choice(travel_Bills,n),
    "entertainment": np.random.choice(entertainment,n),
    "clothes_and_shopping": np.random.choice(clothes_and_shopping,n),
    "medicine_spending": np.random.choice(medicine_spending,n),
    "other_things": np.random.choice(other_things,n),
    "catering": np.random.choice(catering,n),
    "type_of_room": np.random.choice(type_of_room,n),
    "rent": np.random.choice(rent,n),
    "manage": np.random.choice(manage,n),
    "sharing": np.random.choice(sharing,n),
}
# data
# studentdata
```

```python
df = pd.DataFrame(studentdata) #creating the generated data as pandas data frame
# saving the dataframe as a csv file
df.to_csv('studentsimulated_data.csv')
```
